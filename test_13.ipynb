<h1> 붓꽃 아이리스 데이터 </h1>

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
print(iris.DESCR)
```

```python
iris_df= pd.DataFrame(iris.data, columns = iris.feature_names)
```

```python
iris_df
```

```python
iris_df['diagnosis']=iris.target
```

```python
iris_df.head()
```

```python
iris_df.shape
```

```python
iris_df.info()
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
```

```python
iris_scaled = scalar.fit_transform(iris.data)
```

```python
print(iris.data[0])
```

```python
print(iris_scaled[0])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
Y = iris_df['diagnosis']
X = iris_scaled
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state =0)
```

```python
lr_iris = LogisticRegression()
```

```python
lr_iris.fit(X_train, Y_train)
```

```python
Y_predict = lr_iris.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score
```

```python
confusion_matrix(Y_test, Y_predict)
```

```python
accuracy = accuracy_score(Y_test, Y_predict)
```

```python
print("정확도 :{0: 3.f}", format (accuracy))
```

```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns 
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import re
import time
```

```python
driver = webdriver.Chrome()
```

```python
url = "https://map.kakao.com/"
```

```python
driver.get(url)
```

```python
views = driver.find_element(By.XPATH, "//*[@id='search.keyword.query']")
```

```python
views.send_keys("냉면맛집")
```

```python
viewsbutton = driver.find_element(By.XPATH, "//*[@id='search.keyword.submit']")
```

```python
driver.execute_script("arguments[0].click();", viewsbutton)
```

```python
html = driver.page_source
```

```python
soup=BeautifulSoup(html, "html.parser")
```

```python
soup
```

```python
moreviews = soup.find_all("a", attrs = {'class' : 'moreview'})
```

```python
page_urls =[]
for moreview in moreviews:
        page_url = moreview.get('href')
        print(page_url)
        page_urls.append(page_url)
```

```python
driver.close()
```

```python
columns = ['score', 'review']
df = pd.DataFrame(columns=columns)
```

```python
driver = webdriver.Chrome()
```

```python
for page_url in page_urls:
   driver.get(page_url)
   time.sleep(2)
   html=driver.page_source
   soup=BeautifulSoup(html, "html.parser")
   contents_div=soup.find("div", attrs={'class': 'cont_mapdetail'})
   rates=contents_div.find_all('span', attrs={'class': 'inner_star'}) # 별점
   reviews=contents_div.find_all('strong', attrs={'class': 'tit_story'}) # 리뷰
   for rate , review in zip(rates, reviews):
       style = rate.get('style')
       width = re.search(r'width:\s*([0-9]+)%', style)
       if width:
           star_rating = int(width.group(1)) / 20  # 별점은 0-100%를 0-5로 변환
       else:
           star_rating = 0
       row=[star_rating, review.text]
       series=pd.Series(row, index=columns)
       df.loc[len(df)] = row
```
