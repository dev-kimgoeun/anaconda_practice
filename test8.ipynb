```python
state_data=pd.read_csv('서울교통공사_1_8호선 역사 좌표(위경도) 정보_20231031 (1).csv', encoding='cp949')
```

```python
state_data
```

```python
map_state = folium.Map(location=[37.5534378, 127.0725414], zoom_start= 15)
for i, state in state_data.iterrows():
    if state['호선'] == 1 :
        folium.Marker(location=[state["위도"], state["경도"]], popup=state["역명"], icon=folium.Icon(color="red", icon="star"))
```

```python
map_state.save('map_state.html')
```
